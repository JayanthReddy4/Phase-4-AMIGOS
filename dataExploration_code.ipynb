```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv('Full_data.csv')
# we will use the above dataset to find and remove the outliers
print(dataset.shape)  # We see that dataset has 1460 rows and 81 columns.
dataset[["rating", "price_level"]]
# notice the outliers or seperate dots from above 4000 on x-axis
sns.jointplot(x=dataset['rating'], y=dataset['price_level'])

```

```python
# create box plot of only one variable as of now to demonstrate
# You need to clearly examine each graph and try to remove the outliers from it. 


def boxplot(x,y,**kwargs):
            sns.boxplot(x=x,y=y)
            x = plt.xticks(rotation=90)

p = pd.melt(dataset, id_vars='user_ratings_total', value_vars='MSZoning')
g = sns.FacetGrid (p, col='variable', col_wrap=2, sharex=False, sharey=False, size=5)
g = g.map(boxplot, 'value','user_ratings_total')
g
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset = pd.read_csv('Full_data.csv')
numeric_data = dataset.select_dtypes(include=[np.number])
categorical_data = dataset.select_dtypes(exclude=[np.number])
corr = numeric_data.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr)
print (corr['price_level'].sort_values(ascending=False)[:10]) #top 10 correlations
#print (corr['price_level'].sort_values(ascending=False)[-5:]) #least 5 correlations
```
